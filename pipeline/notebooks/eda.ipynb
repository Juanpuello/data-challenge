```python
import pandas as pd
import requests
import numpy as np
```

```python
response_apfel_subscriptions = requests.get("http://localhost:5050/apfel/subscriptions")
print(response_apfel_subscriptions.status_code)
if response_apfel_subscriptions.status_code == 200:
    print("Request successful!")
else:
    msg = f"Request failed with status code: {response_apfel_subscriptions.status_code}"
    print(msg)
    raise Exception(msg)
df_apfel_subscriptions = pd.DataFrame(response_apfel_subscriptions.json()['events'])
print('Apel Subs shape: ', df_apfel_subscriptions.shape)
print('Apel Subs columns: ', df_apfel_subscriptions.iloc[0])
df_apfel_subscriptions.head()
```

```python
df_apfel_subscriptions.info()
```

```python
df_apfel_subscriptions.describe()
```

```python
df_apfel_subscriptions.replace('', np.nan, inplace=True)
date_columns = ['customer_created_at', 'event_timestamp']
for col in date_columns:
    df_apfel_subscriptions[col] = pd.to_datetime(df_apfel_subscriptions[col], errors='raise')
numeric_columns = ['amount', 'tax_amount']
for col in numeric_columns:
    df_apfel_subscriptions[col] = pd.to_numeric(df_apfel_subscriptions[col], errors='raise')
df_apfel_subscriptions.info()
```

```python
df_apfel_subscriptions.describe()
```

```python
df_apfel_subscriptions.nunique()
```

```python
df_id_agg = df_apfel_subscriptions.groupby('session_id').agg(
    unique_ids=('session_id', 'count'),
).reset_index()
duplicate_session_ids = df_id_agg[df_id_agg['unique_ids'] > 1]['session_id']
duplicate_session_rows = df_apfel_subscriptions[df_apfel_subscriptions['session_id'].isin(duplicate_session_ids)]
duplicate_session_rows
```

```python
df_id_agg = df_apfel_subscriptions.groupby('customer_uuid').agg(
    unique_ids=('customer_uuid', 'count'),
).reset_index()
duplicate_cust_ids = df_id_agg[df_id_agg['unique_ids'] > 2]['customer_uuid']
duplicate_cust_rows = df_apfel_subscriptions[df_apfel_subscriptions['customer_uuid'].isin(duplicate_cust_ids)].sort_values('customer_uuid')
duplicate_cust_rows
```

```python
df_id_agg = df_apfel_subscriptions.groupby('event_id').agg(
    unique_ids=('event_id', 'count'),
).reset_index()
duplicate_event_ids = df_id_agg[df_id_agg['unique_ids'] > 1]['event_id']
duplicate_event_rows = df_apfel_subscriptions[df_apfel_subscriptions['event_id'].isin(duplicate_event_ids)].sort_values('event_id')
duplicate_event_rows.drop_duplicates()
```

```python
url = "http://localhost:5050/fenster/subscriptions"
df_fenster_subscriptions = pd.read_csv(url)
print('Fenster Subs shape: ', df_fenster_subscriptions.shape)
print('Fenster Subs columns: ', df_fenster_subscriptions.iloc[0])
df_fenster_subscriptions.head()
```

```python
df_fenster_subscriptions.info()
```

```python
def make_utc(row, column):
    return pd.Timestamp(row[column]).tz_localize(row['tz_clean']).tz_convert('UTC')

df_fenster_subscriptions.replace('', np.nan, inplace=True)
tz_map = {
    'AEST': 'Australia/Sydney',
    'GMT':  'UTC',
    'EST':  'America/New_York',
    'CET':  'Europe/Berlin',
    'PST':  'America/Los_Angeles',
    'CST':  'America/Chicago',
    'JST':  'Asia/Tokyo',
}
df_fenster_subscriptions['tz_clean'] = df_fenster_subscriptions['tz'].map(tz_map).fillna(df_fenster_subscriptions['tz'])
date_columns = ['ts', 'signup_ts']
for col in date_columns:
    df_fenster_subscriptions[col] = pd.to_datetime(df_fenster_subscriptions[col], errors='raise')
    df_fenster_subscriptions[col] = df_fenster_subscriptions.apply(make_utc, axis=1, column=col)
numeric_columns = ['price', 'tax']
for col in numeric_columns:
    df_fenster_subscriptions[col] = pd.to_numeric(df_fenster_subscriptions[col], errors='raise')

df_fenster_subscriptions.info()
```

```python
df_fenster_subscriptions.nunique()
```

```python
df_id_agg = df_fenster_subscriptions.groupby('id').agg(
    unique_ids=('row_hash', 'count'),
).reset_index()
duplicate_ids = df_id_agg[df_id_agg['unique_ids'] > 1]['id']
duplicate_rows = df_fenster_subscriptions[df_fenster_subscriptions['id'].isin(duplicate_ids)]
duplicate_rows
```
